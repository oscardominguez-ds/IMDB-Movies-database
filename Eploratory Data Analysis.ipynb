```python
import pandas as pd
#Concat resource - https://pandas.pydata.org/docs/user_guide/merging.html

df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
df_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)

frames = [df_2000, df_2001]

result = pd.concat(frames)
result.head()
```

```python
result.info()
```

# How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
- Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

```python
#Creating filter for budget then filering dataframe to get values
filter1 = result['budget'] > 0 
result_filtered = result[filter1]
print(f'There are a total of {result_filtered.shape[0]} movies that had valid financial for budget')
```

```python
#Creating filter for revenue then filering dataframe to get values
filter2 = result['revenue'] > 0 
result_filtered1 = result[filter2]
print(f'There are a total of {result_filtered1.shape[0]} movies that had valid financial for budget')
```

# How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
result['certification'].value_counts()
```

# 3. What is the average revenue per certification category?

```python
#Using group 
revenue_per_certification = result.groupby('certification')['revenue'].mean()
revenue_per_certification
```

# What is the average budget per certification category?

```python
budget_per_certification = result.groupby('certification')['budget'].mean()
budget_per_certification
```
